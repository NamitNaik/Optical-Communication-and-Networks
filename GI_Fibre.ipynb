```
import numpy as np

import matplotlib.pyplot as plt

a=50
a1=125

n1=1.457

r=np.linspace(0,a1,a1+1)
n=np.linspace(0,a1,a1+1)

delta1=0.001

for alpha1 in range(1,10):
  for i in range(len(r)):
    if i<a:
      m=2*delta1*(pow((i/a),alpha1))
      n[i]=float(n1*(pow((1-m),0.5)))
    else:
      m=2*delta1
      n[i]=float(n1*(pow((1-m),0.5)))
  plt.plot(r,n)

plt.xlabel("Variable Defining The Optical Fibre Radius, r---> ")
plt.ylabel("Refractive Index, n(r)---> ")
plt.title("Refractive Index Profile")
plt.show()
```
