```
import numpy as np
import matplotlib.pyplot as plt

h=6.26e-34
c=3e8
lam=870e-9
tr=60e-9
tnr=100e-9
q=1.6e-19
t=(tr*tnr)/(tr+tnr)
print(t)

nint=t/tr
print(nint)

I=np.arange(0,40,1)*1e-3
pint=[0]*I

for i in range(40):
  pint[i]=(nint*I[i]*h*c)/(q*lam)

plt.plot(I,pint)
plt.title('Characteristic Curve For Led')
plt.xlabel('Current(A)--------->')
plt.ylabel('Internally Generated Optical Power(W)--------->')
plt.grid()
plt.show()

```
