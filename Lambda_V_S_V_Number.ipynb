```
import numpy as np
import matplotlib.pyplot as plt

n1=1.46
n2=1.455

NA=pow(pow(n1,2)-pow(n2,2),0.5)
print(NA)

d=8e-06

lambda1=np.arange(0.8,1.6,0.01)
v=np.arange(0.8,1.6,0.01)

for i in range(len(lambda1)):
  v[i]=float((np.pi*d*NA)/(lambda1[i]*1e-06))

plt.figure(figsize=(10,6))
plt.plot(lambda1,v) 
plt.xlabel('Wavelength (micrometer)_Namit Naik') 
plt.ylabel('V-Number')
plt.title('Wavlength V/S V-Number for d=8 micrometer')
plt.axhspan(min(v),2.405,color='r', alpha=0.1)
plt.axvspan(1.26,1.6,alpha=0.2)
plt.annotate('Single Mode Operation',xy=(1.29,2), color='black')
plt.show()
  
```
