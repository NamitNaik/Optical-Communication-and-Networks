```
import numpy as np
import matplotlib.pyplot as plt

lambda1=850
lambda2=1300
c=3e05
sig_lam=np.arange(10,100,10)
sig_m1=np.float16(np.arange(10,100,10))
sig_m2=np.float16(np.arange(10,100,10))

factor=0.025

M1=factor/(c*lambda1)
M2=factor/(c*lambda2)

for i in range(len(sig_lam)):
  sig_m1[i]=M1*sig_lam[i]*1e12
  sig_m2[i]=M2*sig_lam[i]*1e12

plt.plot(sig_lam,sig_m1)
plt.plot(sig_lam,sig_m2)

plt.xlabel('Spectral Width Of Light Source(nm)--->')
plt.ylabel('RMS Pulse Broadening[ps/(nm-km)]--->')
plt.title('RMS Pulse Broadening Due To Material Dispersion')

plt.grid()
plt.legend(["Peak Wavelength = 850 nm","Peak Wavelength = 1300 nm"], loc="upper left")
plt.show()

```
