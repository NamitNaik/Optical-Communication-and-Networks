```
import numpy as np

import matplotlib.pyplot as plt

a=50
a1=125

n1=1.457
n2=1.45

r=np.linspace(0,a1,a1+1)
n=np.linspace(0,a1,a1+1)

for i in range(len(r)):
  if i<=50:
    n[i]=n1
  else:
    n[i]=n2

plt.plot(r,n)
plt.xlabel("Variable Defining The Optical Fibre Radius, r---> ")
plt.ylabel("Refractive Index, n(r)---> ")
plt.title("Refractive Index Profile")
plt.show()

```
